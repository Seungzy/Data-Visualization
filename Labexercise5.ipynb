```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('MegaGymDataset.csv')
```

```python
print(df.head())
print(df['Type'].unique())
print(df['Type'].nunique())
print(df['Type'].value_counts()['Strength'])
print((df['Type']=='Strength').sum())
print(df['Type'].count())
print(df['Equipment'].unique())
print(df['BodyPart'].unique())
print(df['Level'].unique())
print(df.loc[df['Level']=='Intermediate', 'Rating'].mean())
print(df['Rating'].mean())
print(df['Rating'].median())
print(df['Rating'].mode()[0])
print(df['Rating'].sum())
print(df['Rating'].min())
print(df['Rating'].max())
```

```python
bar_level_label = np.array(['Beginner','Intermediate','Expert'])

bar_level_mean = np.array((
    df.loc[df['Level']=='Beginner','Rating'].mean(),
    df.loc[df['Level']=='Intermediate','Rating'].mean(),
    df.loc[df['Level']=='Expert','Rating'].mean()
))
```

```python
plt.bar(bar_level_label, bar_level_mean)
plt.title("Average Rating per Exercise Level")
plt.show()
```

```python
#Line Plot Q1
rating_by_level = df.groupby('Level')['Rating'].mean()

plt.plot(rating_by_level)
plt.title("Average Rating by Level")
plt.xlabel("Level")
plt.ylabel("Rating")
plt.show()
```

```python
#Line Plot Q2
level_counts = df['Level'].value_counts()

plt.plot(level_counts.index, level_counts.values, marker='o')
plt.title("Number of Exercises per Level")
plt.xlabel("Exercise Level")
plt.ylabel("Count")
plt.show()
```

```python
#Line Plot Q3
rating_by_bodypart = df.groupby('BodyPart')['Rating'].mean()

plt.plot(rating_by_bodypart.index, rating_by_bodypart.values, marker='o')
plt.xticks(rotation=90)
plt.title("Average Rating per Body Part")
plt.xlabel("Body Part")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Line Plot Q4
equipment_counts = df['Equipment'].value_counts()

plt.plot(equipment_counts.index, equipment_counts.values, marker='o')
plt.xticks(rotation=90)
plt.title("Number of Exercises per Equipment")
plt.xlabel("Equipment")
plt.ylabel("Count")
plt.show()
```

```python
#Line Plot Q5
rating_by_type = df.groupby('Type')['Rating'].mean()

plt.plot(rating_by_type.index, rating_by_type.values, marker='o')
plt.xticks(rotation=90)
plt.title("Average Rating per Exercise Type")
plt.xlabel("Exercise Type")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Scatter Plot Q1
level_codes = df['Level'].astype('category').cat.codes

plt.scatter(level_codes, df['Rating'])
plt.title("Exercise Level vs Rating")
plt.xlabel("Level Code")
plt.ylabel("Rating")
plt.show()
```

```python
#Scatter Plot Q2
type_codes = df['Type'].astype('category').cat.codes

plt.scatter(type_codes, df['Rating'])
plt.title("Exercise Type vs Rating")
plt.xlabel("Type Code")
plt.ylabel("Rating")
plt.show()
```

```python
#Scatter Plot Q3
bodypart_codes = df['BodyPart'].astype('category').cat.codes

plt.scatter(bodypart_codes, df['Rating'])
plt.title("Body Part vs Rating")
plt.xlabel("Body Part Code")
plt.ylabel("Rating")
plt.show()
```

```python
#Scatter Plot Q4
equipment_codes = df['Equipment'].astype('category').cat.codes

plt.scatter(equipment_codes, df['Rating'])
plt.title("Equipment Type vs Rating")
plt.xlabel("Equipment Code")
plt.ylabel("Rating")
plt.show()
```

```python
#Scatter Plot Q5
exercise_number = range(len(df))

plt.scatter(exercise_number, df['Rating'])
plt.title("Exercise Index vs Rating")
plt.xlabel("Exercise Index")
plt.ylabel("Rating")
plt.show()
```
