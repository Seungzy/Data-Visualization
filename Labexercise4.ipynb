```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("data.csv")
print(df.head())
```

```python
plt.figure()

plt.scatter(df["Pulse"], df["Calories"])

plt.title("Pulse vs Calories")
plt.xlabel("Pulse")
plt.ylabel("Calories")

plt.show()
```

```python
bar_data = df[["Date", "Calories"]].dropna()
bar_data["Date"] = bar_data["Date"].astype(str)

plt.figure()
plt.bar(bar_data["Date"], bar_data["Calories"])
plt.title("Calories per Date")
plt.xlabel("Date")
plt.ylabel("Calories")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure()

plt.pie(df["Calories"][:5], labels=df["Date"][:5], autopct="%1.1f%%")

plt.title("Calories Distribution (First 5 Records)")

plt.show()
```

```python
line_data = df[["Date", "Calories"]].dropna()

line_data["Date"] = line_data["Date"].astype(str)

plt.figure()

plt.plot(line_data["Date"], line_data["Calories"], marker="o")

plt.title("Calories Over Time")
plt.xlabel("Date")
plt.ylabel("Calories")

plt.xticks(rotation=90)

plt.show()
```
